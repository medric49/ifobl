```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.signal import savgol_filter
import numpy as np
```

## Reacher Hard, episode return

```python
ifo_1 = pd.read_csv('experiments/reacher_hard2/ifobl/1.csv')

ct_1 = pd.read_csv('experiments/reacher_hard2/ct/1.csv')
ct_2 = pd.read_csv('experiments/reacher_hard2/ct/2.csv')
```

```python
y = ifo_1['Value']
x = np.arange(0, y.shape[0]) * 1e4
y_ = savgol_filter(y, 51, 3)
color = np.array([0, 0, 128, 50]) / 255.
color_ = color.copy()
color_[-1] = 1.
plt.plot(x, y, color=color)
plt.plot(x, y_, color=color_, label='ours')

y = ct_2['Value']
y_ = savgol_filter(y, 51, 3)
color = np.array([32, 189, 79, 50]) / 255.
color_ = color.copy()
color_[-1] = 1.
plt.plot(x, y, color=color)
plt.plot(x, y_, color=color_, label='CT-10')


y = ct_1['Value']
y_ = savgol_filter(y, 51, 3)
color = np.array([189, 32, 63, 50]) / 255.
color_ = color.copy()
color_[-1] = 1.
plt.plot(x, y, color=color)
plt.plot(x, y_, color=color_, label='CT-1')

plt.legend()
plt.ylim([0, 20])
plt.grid(True)
plt.xlabel('Frames')
plt.ylabel('120-frames episode return')
plt.show()
```
